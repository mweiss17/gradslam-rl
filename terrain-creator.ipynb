```python
res = 201
```

```python
def pyramid(n):
    r = np.arange(n)
    d = np.minimum(r, r[::-1])
    return np.minimum.outer(d, d)
```

```python
# Make the cone
scale_factor = 4
pyramid = pyramid(res)
bias = pyramid.max() - 5
cone_terrain = np.round(scale_factor * (bias - pyramid).clip(min=0) / pyramid.max(), 1)
```
